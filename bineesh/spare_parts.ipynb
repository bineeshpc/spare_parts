```python
def total_percentage_of_missing_values(data):
    """ This formula works for all data
    percentage of values 
    """
    return (data.shape[0] - data.dropna().shape[0]) / data.shape[0] * 100
```

```python
data1 = pd.get_dummies(data, columns=['INVOICE LINE TEXT'])
```

```python
data2 = data1[['Invoice Date', 'Job Card Date', 'Business Partner Name', 'Vehicle No.',
       'Vehicle Model', 'Current KM Reading', 'INVOICE LINE TEXT_ENGINE OIL']]
```

```python
data2.groupby('Invoice Date').agg({'INVOICE LINE TEXT_ENGINE OIL': 'sum'})
```

```python
import datetime

datetime.datetime.strptime('2019-01-01', '%Y-%m-%d')
```

```python
year_month = data2['Invoice Date'].apply(lambda x: datetime.datetime.strptime(f'{x.year}-{x.month}-01', '%Y-%m-%d'))
```

```python
data2['year_month'] = year_month
```

```python
data2.head()
```

```python
data3 = data2.groupby('year_month').agg({'INVOICE LINE TEXT_ENGINE OIL': 'sum'})
```

```python
dt_fmt = '%Y-%m-%d'
data4 = data3.loc[(data3.index > datetime.datetime.strptime('2017-05-01', dt_fmt)) & (data3.index < datetime.datetime.strptime('2019-01-01', dt_fmt))]
```

```python
data4.index
```

```python
data4
```

```python
from statsmodels.tsa.arima_model import ARIMA
import itertools
```

```python
import sys
sys.maxsize
```

```python
r = range(0, 2)
min_aic = sys.maxsize

for p, d, q in itertools.product(r, r, r):
    try:
        model_arima = ARIMA(data4, order=(p, d, q))
        model_arima_fit = model_arima.fit()
        pred = model_arima_fit.forecast(steps=9)[0]
        print(p, d, q, pred, model_arima_fit.aic)
        if model_arima_fit.aic < min_aic:
            result = (p, d, q, pred, model_arima_fit.aic)
            min_aic = model_arima_fit.aic
    except:
        print(p, d, q, 'bad')
        
```
