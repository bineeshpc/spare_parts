```python
year_month = data1['Invoice Date'].apply(lambda x: datetime.datetime.strptime(f'{x.year}-{x.month}-01', '%Y-%m-%d'))
```

```python
data1['year_month'] = year_month
```

```python
data1.head()
```

```python
data3 = data1.groupby('year_month').agg({i: 'sum' for i in invoice_line_info})
```

```python
dt_fmt = '%Y-%m-%d'
data4 = data3.loc[(data3.index >= datetime.datetime.strptime('2017-06-01', dt_fmt)) & (data3.index < datetime.datetime.strptime('2019-01-01', dt_fmt))]
```

```python
data4.index
```

```python
from statsmodels.tsa.arima_model import ARIMA
import itertools
```

```python
import sys
sys.maxsize
```

```python
models = {}
for invoice_line in invoice_line_info:
    data5 = data4[invoice_line]
    r = range(0, 2)
    min_aic = sys.maxsize

    for p, d, q in itertools.product(r, r, r):
        try:
            model_arima = ARIMA(data5, order=(p, d, q))
            model_arima_fit = model_arima.fit()
            pred = model_arima_fit.forecast(steps=9)[0]
            print(p, d, q, pred, model_arima_fit.aic)
            if model_arima_fit.aic < min_aic:
                result = (p, d, q, pred, model_arima_fit.aic)
                min_aic = model_arima_fit.aic
        except:
            print(p, d, q, 'bad')
    print(invoice_line, result)
    models[invoice_line] = (model_arima, model_arima_fit)      
```

```python
models
```

```python
for invoice_line in invoice_line_info:
    next_month = models[invoice_line][1].forecast(steps=1)[0]
    print(invoice_line, next_month)
```

```python
picklelib.dump(models, 'spare_parts.pickle')
```
